```python
columns = {"ID": (0,11),
           "LATITUDE": (12, 20), 
           "LONGITUDE": (21, 30),
           "ELEVATION": (31, 37),
           "STATE": (38, 40),
           "NAME": (41, 71),
           "GSN FLAG": (72, 75),
           "HCN/CRN FLAG": (76, 79),
           "WMO ID": (80, 85)}
```

```python
with open("../data/ghcnd-stations.txt") as f:
    lines = f.readlines()
```

```python
rows = []
for line in lines:
    row = dict()
    for k, (start, stop) in columns.items():
        row[k] = line[start:stop]
    rows.append(row)
```

```python
df = pd.DataFrame(rows, columns = columns.keys())
```

```python
url = "https://www.weather.gov/mdl/mos_stations_stadrggfs2009"
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text)
```

```python
stations = soup.find_all('pre')
```

```python
columns_mos = {'ID':(0,4),
           'NAME':(4,24),
           'STATE':(24,30),
           'LATITUDE': (31,40),
           'LONGITUDE':(41,50)}
```

```python
rows = []
for station in stations:
    for s in station.text.split("\r\n"):   
        if not s:
            continue
        row = dict()
        for key, (start, stop) in columns_mos.items():
            row[key] = s[start:stop].strip()
        rows.append(row)
```

```python
mosdf = pd.DataFrame(rows)
```

```python
mosdf[mosdf.any(axis=1)].to_csv("mos_stations.csv")
```
